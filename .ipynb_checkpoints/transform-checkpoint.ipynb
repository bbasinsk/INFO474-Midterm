```python
import pandas as pd 
import numpy as np 
```

```python
data = pd.read_csv('SimpsonsData.csv')
```

```python
data.drop(["Bar Size", "Directed by", "Guest star", "No.", "Synopsis", "Title", "Written by", "Role(s)", "Production Code (Characters)", "Production Code"], axis=1)
```

For each season

 - year / season #
 - avg viewers (in millions)
 - number of episodes
 - most watched episode
 - viewers of most watched

```python
aggregation = {
    'Season': {
        'year': lambda x: max(x) + 1989
    },
    'U.S. Viewers': {
        'most_viewers': 'max',
        'avg_viewers': 'mean'
    }
}

data.groupby('Season').agg(aggregation)
```
